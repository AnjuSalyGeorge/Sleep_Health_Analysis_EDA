Sleep Health and Lifestyle Dataset comprises 400 rows and 13 columns, covering a wide range of variables related to sleep and daily habits. It includes details such as gender, age, occupation, sleep duration, quality of sleep, physical activity level, stress levels, BMI category, blood pressure, heart rate, daily steps, and the presence or absence of sleep disorders in order to identify the relationship between sleep health and life style with productivity.<br><br><br>

⏹<font size="5"><b>Key Features Of data set</b></font><br>
Comprehensive Sleep Metrics: Explore sleep duration, quality, and factors influencing sleep patterns.<br>
Lifestyle Factors: Analyze physical activity levels, stress levels, and BMI categories.<br>
Sleep Disorder Analysis: Identify the occurrence of sleep disorders such as Insomnia and Sleep Apnea.

⏹<font size="5"><b>Dataset Columns</b></font><br>
Person ID: An identifier for each individual.<br>
Gender: The gender of the person (Male/Female).<br>
Age: The age of the person in years.<br>
Occupation: The occupation or profession of the person.<br>
Sleep Duration (hours): The number of hours the person sleeps per day.<br>
Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.<br>
Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.<br>
Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.<br>
BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).<br>
Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.<br>
Heart Rate (bpm): The resting heart rate of the person in beats per minute.
Daily Steps: The number of steps the person takes per day.<br>
Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).<br>

<font size="3"><b>Sleep Disorder Column</b></font><br>
None: The individual does not exhibit any specific sleep disorder.<br>
Insomnia: The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.<br>
Sleep Apnea: The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.

⏹<font size="5"><b>Dataset Retrieval</b></font><br>
We read the dataset from our GitHub repository and load the modules we want throughout this report. We also display head and tail of the Dataset.


```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import io
import requests
import matplotlib.pyplot as plt
import seaborn as sns

#Import the dataset from the URL
df_url = 'https://raw.githubusercontent.com/anjugeorgeclt/statanjugeorge_clt/main/Sleep_health_and_lifestyle_dataset.csv'
url_content = requests.get(df_url, verify=False).content

#Read the imported CSV data into a DataFrame
sleeph = pd.read_csv(io.StringIO(url_content.decode('utf-8')))
sleeph.shape  #Display number of rows and columns
```

```python
sleeph.tail()  #Display last five rows form the dataset
```

Ages are somewhat spread out here, which represents a relatively large range of ages in our dataset. Also, it does not show any visible extreme values that can skew our values based on age.

⏹<font size="3"><b>2A: Plot Displaying Single Categorical Variable</b></font><br>

```python
# Count the occurrences of each sleep disorder in the 'sleep_order' column
occupation_counts = sleeph['sleep_disorder'].value_counts()

# Create a figure with desired size
plt.figure(figsize=(12, 5))

# Define a list of colors
colors = ['pink', 'yellow', 'orange']

# Plot the bar chart with bars and custom colors
occupation_counts.plot(kind='bar', stacked=True,color=colors)

plt.title('Figure 2A:Bar Plot of sleep disorder',fontsize='18')
plt.xlabel('Sleep Disorder')
plt.ylabel('Count')
caption='''Above Bar plot counts occurences of individuals in each Sleep disorder category'''
plt.figtext(0.5, -0.17, caption, wrap=True, horizontalalignment='center', fontsize=12) #Caption for the plot
plt.show()
```

The above Bar plot visualizes the three values of 'sleep_ disorder' column mentioned in the dataset. The values here are 'None','Sleep Apnea' and 'Insomnia'.The bar chart with three bars represents number of occurences of a specific category of sleep disorder.The plot clearly depicts that the count of people in our dataset having any sleep disorder is low compared to people having no sleep disorder.

For Sleep Apnea 50% of people have very high Physical Activity Level.

⏹<font size="3"><b>4A: Plots Displaying relationship between two variables</b></font><br>

```python

# Stacked bar plot of sleep disorder, grouped by bmi_category and count the occurrences

sleeph.groupby(['sleep_disorder', 'bmi_category']).size().unstack().plot(kind='bar', stacked=True)

# Add title and labels
plt.title('Figure 4A: Stacked Bar Plot of Sleep Disorder and BMI Category',fontsize='15')
plt.xlabel('Sleep Disorder')
plt.ylabel('Count')
plt.tight_layout()

caption='''Above Stacked Bar plot shows the relationship between Sleep Disorder and BMI'''
plt.figtext(0.5, -0.12, caption, wrap=True, horizontalalignment='center', fontsize=12)
# Display the plot

plt.show()
```

The above Stacked Bar plot clearly visualizes the relationship between two categorical variables; Sleep disorder and BMI. From the plot, most people from our dataset didn't have any sleep disorders.The disorders can be related to BMI value of individuals such that People having 'Insomnia' and 'Sleep Apnea' are mostly 'overweight' . Among the people having 'No Disorder' , the number of occurences of individuals having 'Normal weight' is higher followed by 'Overweight' and 'Under weight', but no one is from 'obese' category. Thus, the chart shows that individuals who are 'Over weight' and 'Obese' are more likely to experience any sleep disorders, particularly 'Insomnia' and 'Sleep Apnea'.

⏹<font size="3"><b>4B: Plots Displaying relationship between two variables</b></font><br>

```python
# scatterplot of stress level and quality of sleep columns in the sleeph dataframe+
plt.figure(figsize=(15,8))
plt.scatter(sleeph['stress_level'],sleeph['quality_of_sleep'],c=sleeph['quality_of_sleep'],cmap='YlGnBu_r',alpha=0.3)
plt.title("Figure 4B: Scatterplot of Stress level and Quality of sleep",fontsize=15)
plt.xlabel('Stress level')
plt.ylabel('Quality of sleep')
#To plot Best Fit line
coefficients = np.polyfit(sleeph['stress_level'],sleeph['quality_of_sleep'], 1)
line_y = np.polyval(coefficients,sleeph['stress_level'])
plt.plot(sleeph['stress_level'],line_y,color='r',linestyle='--',lw=1)
caption='''Above Scatterplot shows the relationship between Stress level and Quality of sleep'''
plt.figtext(0.5, -0.02, caption, wrap=True, horizontalalignment='center', fontsize=12)
plt.show()
```

The scatterplot of stress level and quality of sleep shows as quality of sleep decreases there is an increase in stress level and the line of best fit depicts the relationship exactly.So it can be stated that stress is an important factor that influences the quality of sleep and by decreasing stress levels the participants can achieve a better quality of sleep.




<font size="5"><b>Three Variable Visualization</b></font><br>

⏹<font size="3"><b>5: Plots Displaying information about 3 or more variable</b></font><br>

```python
plt.figure(figsize=(12,5))
#Plot Scatter plot of continues variable age and catogorical variables BMI category and sleep disorder
sns.scatterplot(x=sleeph['age'],y=sleeph['sleep_disorder'],hue=sleeph['bmi_category'],s=150)#
plt.xlabel('Age', fontsize=15)
plt.ylabel('Sleep Disorder', fontsize=15)
plt.title('Figure 5A: Scatterplot of sleep disorder down by age coloured by BMI',fontsize=18)
#Add legend indicating BMI categories on the lower right side
plt.legend(loc='lower right')
#Adding caption to the plot
caption2='''Above Scatter plot shows the relationship between age of people,BMI category and sleep disorder'''
plt.figtext(0.5, -0.02, caption2, wrap=True, horizontalalignment='center', fontsize=12)
plt.show()
```

From the scatter plot, we can see a positive correlation between age and sleep disorder, which means older individuals are more likely to have a sleep disorder than younger individuals.Also,individuals with a higher BMI category are mostly affected by sleep disorders, and most obese people have a higher probability of having a sleep disorder.Most overweight individuals suffer from sleep apnea.Overall, the scatterplot shows that the combination of older age and a higher BMI category is associated with an even greater likelihood of reporting any sleep disorder. This suggests that age and BMI category may have a cumulative effect on sleep health.
